```python
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = open(r"PATH","r", encoding = 'utf-8')
```

```python
data = f.read()
print(data)
```

```python
# pattern =  r'\d{1,2}/\d{1,2}/\d{2},\d{1,2}:\d{2}\s*(AM|PM)?\s*-\s*(.*?):\s*(.*)' pattern for whole 
# pattern = r"\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s*(AM|PM)?\s-\s" #pattern till date and time
# pattern =  r"\d{1,2}/\d{1,2}/\d{2},\s\d{1,2}:\d{2}\s*(?:AM|PM)?\s*-\s[^-]+"
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s(?:AM|PM)?\s-\s'
```

```python
message = re.split(pattern, data)[1:]
message
```

```python
dates = re.findall(pattern, data)
```

```python
dates = [s.replace('\u202F', ' ')for s in dates] # we had to run this as '/s' in pattern caught non-breaking element "\u202f" need to be replace by space " "
```

```python
dates
```

```python
df = pd.DataFrame({'user_message' : message, 'message_date': dates})
```

```python
df.info()
```

```python
#converting dates into date format from strings (object)

df['message_date']= pd.to_datetime(df['message_date'], format = '%m/%d/%y, %I:%M %p - ')
df.rename(columns={'message_date':'date'}, inplace = 1)
```

```python
df.head()
```

```python
df.shape
```

```python
users= []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
df['user']=users
df['message']= messages
df.drop(columns=['user_message'],inplace=True)
```

```python
from collections import Counter

def word_counter(user, df):

    words=[]

    if user != 'overall':
        df = df[df['user']==user]


    temp = df[df['user']!='group_notification']
    temp = temp[temp['message'] != '<Media omitted>\n']

    f = open("HinglishStopWordFile\stop_hinglish.txt", 'r')
    stopWords = f.read()
    for msg in temp['message']:
        for word in msg.lower().split():
            if word not in stopWords:
                words.append(word)
        
    

    return pd.DataFrame(Counter(words).most_common(20))
```

```python
temp = word_counter("overall", df)
```
